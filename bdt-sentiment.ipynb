```python
documents = [
    ("D1", "the cat sat on the mat"),
    ("D2", "the dog sat on the log"),
]

import multiprocessing

def map_function(document):
    doc_id, text = document
    words = text.split()
    output = []
    for word in words:
        output.append((word, doc_id))
    return output

# Simulate parallel mapping
with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
    mapped_data = pool.map(map_function, documents)

# Flatten the list of key-value pairs
mapped_data = [item for sublist in mapped_data for item in sublist]

def reduce_function(mapped_data):
    inverted_index = {}
    for word, doc_id in mapped_data:
        if word not in inverted_index:
            inverted_index[word] = []
        inverted_index[word].append(doc_id)
    return inverted_index

inverted_index = reduce_function(mapped_data)

for word, doc_ids in inverted_index.items():
    print(f"{word}: {doc_ids}")
```

```python
documents = [
    ("https://www.gutenberg.org/files/1112/1112.txt", RJ),
    ("https://www.gutenberg.org/cache/epub/1128/pg1128.txt", KL),
]


import multiprocessing

def map_function(document):
    doc_id, text = document
    words = text.split()
    output = []
    for position, word in enumerate(words):
        output.append((word, (doc_id, position)))
    return output

# Simulate parallel mapping
with multiprocessing.Pool(processes=multiprocessing.cpu_count()) as pool:
    mapped_data = pool.map(map_function, documents)

# Flatten the list of key-value pairs
mapped_data = [item for sublist in mapped_data for item in sublist]

def reduce_function(mapped_data):
    inverted_index = {}
    for word, (doc_id, position) in mapped_data:
        if word not in inverted_index:
            inverted_index[word] = []
        inverted_index[word].append((doc_id, position))
    return inverted_index

inverted_index = reduce_function(mapped_data)

for word, positions in inverted_index.items():
    print(f"{word}: {positions}")
```
